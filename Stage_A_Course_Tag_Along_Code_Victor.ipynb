```python
# load the food csv file dataset into a pandas dataframe

food_df = pd.read_csv("FoodBalanceSheets_E_Africa_NOFLAG.csv", encoding = "latin-1")
```

##### Question 01<br>Select columns 'Y2017' and 'Area', perform a grouby operation on 'Area'.<br> Which of these Areas had the highest sum in 2017?

```python
# filtering the dataset to only include 'Y2017' and 'Area' data columns 
filter_01 = food_df[['Area','Y2017']]
```

```python
#  Groupby 'Area' and sum the 'Y2017' values for each area, return the top 10 areas with the highest 'Y2017' values
filter_01.groupby('Area')['Y2017'].sum().sort_values(ascending=False).head(10)
```

```python
# choose the highest sum area from the above result
highest_sum_area = filter_01.groupby('Area')['Y2017'].sum().idxmax()
print('Area_of_interest =',highest_sum_area)
```

##### Question 02<br>What is the total number of unique countries in the dataset?

```python
# return the number of unique areas in the filtered dataframe
unique_areas = filter_01['Area'].nunique()
print('Unique_Areas_Total =',unique_areas)
```

##### Question 05<br>Perform a groupby operation on ‘Element’<br>What is the total number of the sum of Processing in 2017?

```python
# create a second dataframe that only includes the 'Element' and 'Y2017' data columns
filter_02 = food_df[['Element','Y2017']]
```

```python
# return the processing sum of the 'Y2017'
filter_02.query("Element=='Processing'").groupby('Element')['Y2017'].sum()
```

##### Question 06<br>Which year had the least correlation with ‘Element Code’?

```python
# create a correlation matrix for the columns 'Y2014', 'Y2015', 'Y2016', 'Y2017', and 'Element Code' in the dataframe 'food_df'
corr_matrix = food_df[['Y2014','Y2015','Y2016','Y2017', 'Element Code']].corr()

# find the column with the minimum absolute correlation with the 'Element Code' column and assign it to the variable 'min_corr_feature'
min_corr_feature = corr_matrix['Element Code'].abs().idxmin()

# print the value of 'min_corr_feature'
print('least_correlation_year =',min_corr_feature)
```

##### Question 09<br>What is the total number and percentage of missing data in 2014 to 3 decimal places?

```python
# Count the number of missing values in the 'Y2014' column of the 'food_df' dataframe
missing_values = food_df['Y2014'].isnull().sum()

# Count the total number of values in the 'food_df' dataframe
total_value_count = len(food_df)

# Calculate the percentage of missing values in the 'Y2014' column of the 'food_df' dataframe
missing_percentage = (missing_values/total_value_count)*100
```

```python
# Print the total count of missing values and the percentage of missing values rounded off to 3 decimal places
print('missing_total_count =',missing_values,'\n''missing_percentage =',round(missing_percentage,3))
```

##### Question 11<br>Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the 7th lowest sum in 2017?

```python
# Find the seventh lowest sum of 'Y2017' values by 'Area' 
seventh_lowest_sum = filter_01.groupby('Area')['Y2017'].sum().nsmallest(7).tail(1)

# Print the value of seventh lowest sum 
print(seventh_lowest_sum)
```

##### Question 12<br>What is the mean and standard deviation across the whole dataset for the year 2017 to 2 decimal places?

```python
# return the summary statistics of the first filtered dataframe
filter_01.describe()
```

```python
# Calculate the mean of the 'Y2017' column rounded off to 2 decimal places
avg = round(filter_01['Y2017'].mean(), 2)

# Calculate the standard deviation of the 'Y2017' column rounded off to 2 decimal places
std = round(math.sqrt(filter_01['Y2017'].var()), 2)

# Print the mean and standard deviation of the 'Y2017' column
print('mean =',avg,'\n''standard_deviation =',std)
```

##### Question 14<br>What is the total sum of Wine produced in 2015 and 2018 respectively?<br>Hint: Perform a groupby sum aggregation on 'Item'

```python
# filter out the 'Item', 'Y2015', and 'Y2018' columns from the 'food_df' dataframe and assign it to the variable 'filter_03'
filter_03 = food_df[['Item','Y2015','Y2018']]
```

```python
# return the sum of 'Y2015' and 'Y2018' values for the 'Item' group 'Wine'
filter_03.query("Item == 'Wine'").groupby('Item')['Y2015','Y2018'].sum()
```

##### Question 15<br>What is the total Protein supply quantity in Madagascar 2015?

```python
# filter out the 'Area', 'Element', and 'Y2015' columns from the 'food_df' dataframe and assign it to the variable 'filter_04'
filter_04 = food_df[['Area','Element','Y2015']]

# return the sum of 'Y2015' values for the 'Element' group 'Protein supply quantity (g/capita/day)' and 'Area' group 'Madagascar'
filter_04.query("Area == 'Madagascar'and Element == 'Protein supply quantity (g/capita/day)'").groupby('Element')['Y2015'].sum()
```

##### Question 17<br>Perform a groupby operation on 'Element'.<br>What year has the highest sum of Stock Variation?

```python
# filter out the 'Element', 'Y2014', 'Y2015', 'Y2016', 'Y2017', and 'Y2018' columns from the 'food_df' dataframe and assign it to the variable 'filter_05'
filter_05 = food_df[['Element','Y2014','Y2015','Y2016','Y2017','Y2018']]

# return the sum of 'Y2014', 'Y2015', 'Y2016', 'Y2017', and 'Y2018' values for the 'Element' group 'Stock Variation'
filter_06 = filter_05.query("Element=='Stock Variation'").groupby('Element')['Y2014','Y2015','Y2016','Y2017','Y2018'].sum()

# return the year with the highest sum of 'Stock Variation' values
filter_06.idxmax(axis=1)
```
